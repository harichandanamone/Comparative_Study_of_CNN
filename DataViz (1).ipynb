#T-SNE


### FOR MNIST


```python
import keras
from keras.datasets import mnist
(train_X, train_y), (test_X, test_y) = mnist.load_data()
print('X_train: ' + str(train_X.shape))
print('Y_train: ' + str(train_y.shape))
print('X_test:  '  + str(test_X.shape))
print('Y_test:  '  + str(test_y.shape))
```

```python
from sklearn.manifold import TSNE
data = train_X[0:3000,:]
labels = train_y[0:3000]
model = TSNE(n_components=2, random_state=0)
```

```python
labels
```

```python
nsamples, nx, ny = data.shape
train_dataset = data.reshape((nsamples,nx*ny))
```

```python
tsne_data = model.fit_transform(train_dataset)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
tsne_data = np.vstack((tsne_data.T, labels)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))
```

```python
data_dis = sns.FacetGrid(tsne_df,  hue="label")
data_dis.map(sns.scatterplot, "Dim_1", "Dim_2")
data_dis.add_legend()
```

```python
val = []
# Loop through all label
for i in range(data_f.shape[0]):
    # Extract the label for comparison
    if data_f['label'][i] == 0:
        # Save meaningful label to the results
        val.append('T-shirt/top')
    # Following the same code pattern as the one above
    elif data_f['label'][i] == 1:
        val.append('Trouser')
    elif data_f['label'][i] == 2:
        val.append('Pullover')
    elif data_f['label'][i] == 3:
        val.append('Dress')
    elif data_f['label'][i] == 4:
        val.append('Coat')
    elif data_f['label'][i] == 5:
        val.append('Sandal')
    elif data_f['label'][i] == 6:
        val.append('Shirt')
    elif data_f['label'][i] == 7:
        val.append('Sneaker')
    elif data_f['label'][i] == 8:
        val.append('Bag')
    elif data_f['label'][i] == 9:
        val.append('Ankle boot')
    else:
        print("unexpected label".format(data_f['label'][i]))
     
val
```

```python
data_f['dim 1'] = pca_f[:,0]
data_f['dim 2'] = pca_f[:,1] 
```

```python
data_f.head()
data_f['val'] = val
```
